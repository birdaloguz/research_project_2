```python
##movielens 20m
##pysparnn vs knn with k=2
##recall
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
import json

with open('knn_eval_results.json') as json_file:
    data = json.load(json_file)
    recalls_knn = [data['recalls'][i] for i in data['recalls']]
with open('pysparnn_eval_results_k.json') as json_file:
    data = json.load(json_file)
    recalls_pysparnn = [data['recalls'][i] for i in data['recalls']]
k=[2, 5, 10, 15, 20, 50]

x_labels = k

# Set plot parameters
fig, ax = plt.subplots()
width = 0.1 # width of bar
x = np.arange(6)

ax.bar(x, recalls_knn, width, color='blue', label='knn')
ax.bar(x + width, recalls_pysparnn, width, color='green', label='pysparnn')

ax.set_ylabel('recall@10')
ax.set_ylim(0,0.125)
ax.set_xticks(x + width)
ax.set_xticklabels(x_labels)
ax.set_xlabel('k')
ax.set_title('knn vs pysparnn')
ax.legend()
plt.grid(True, 'major', 'y', ls='--', lw=.5, c='k', alpha=.3)

fig.tight_layout()
plt.show()
%matplotlib inline
```

```python
##movielens 20m
##pysparnn with different #indexes
##recall
import matplotlib.pyplot as plt
import numpy as np

with open('pysparnn_eval_results5.json') as json_file:
    data = json.load(json_file)
    recalls = [data['recalls'][i] for i in data['recalls']]
    ndcg = [data['ndcg'][i] for i in data['ndcg']]
    mrr = [data['mrr'][i] for i in data['mrr']]
x=[2, 5, 10, 15, 20, 50]

fig=plt.figure()
fig.show()
ax=fig.add_subplot(111)

ax.plot(x,recalls,c='r',marker="v",ls='-',label='recall')
ax.plot(x,ndcg,c='m',marker="o",ls='--',label='ndcg',fillstyle='none')
ax.plot(x,mrr,c='k',marker="+",ls=':',label='mrr')
ax.set_xlabel('#index')
plt.legend(loc=2)
plt.draw()
```

```python
##movielens 20m
##different runs with different validation set
from matplotlib import pyplot as plt 
import json
from random import shuffle
with open('knnrange_eval_results.json') as json_file:
    data = json.load(json_file)
    recalls_knn = [data['recalls'][i] for i in data['recalls']]
    
with open('pysparnnrange_eval_results.json') as json_file:
    data = json.load(json_file)
    recalls_pysparnn = [data['recalls'][i] for i in data['recalls']]
    
#generate a list of markers and another of colors 
markers = ["<", ">"]
colors = ['y', 'k']


knn=plt.scatter([i for i in range(10)],recalls_knn,marker=markers[0], color=colors[0])
pysparnn=plt.scatter([i for i in range(10)],recalls_pysparnn,marker=markers[1], color=colors[1])
plt.legend((knn, pysparnn),
           ('knn', 'pysparnn(index=50, cluster=10)'),
           scatterpoints=1,
           loc='upper right')
plt.show() 
```
